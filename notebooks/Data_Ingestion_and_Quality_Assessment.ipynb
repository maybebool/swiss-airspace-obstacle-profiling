```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## 1. Load Dataset
The parquet file was produced by `src/data/fetch_obstacles.py`, which downloads the KMZ from the BAZL STAC API, parses the KML, and saves a GeoDataFrame with WGS84 (EPSG:4326) coordinates.


```python
gdf = gpd.read_parquet('../data/processed/obstacles.parquet')
print(f"Shape: {gdf.shape[0]} rows, {gdf.shape[1]} columns")
print(f"CRS:   {gdf.crs}")
gdf.head()
```

## 2. Schema Overview

```python
pd.DataFrame({
    'dtype': gdf.dtypes,
    'non_null': gdf.notnull().sum(),
    'null': gdf.isnull().sum(),
    'null_pct': (gdf.isnull().sum() / len(gdf) * 100).round(1),
    'n_unique': gdf.nunique()
})
```

## 3. Duplicates

```python
n_dupes = gdf.duplicated(subset='uuid').sum()
print(f"Duplicate UUIDs: {n_dupes}")

if n_dupes > 0:
    print("\nExample duplicates:")
    duped = gdf[gdf.duplicated(subset='uuid', keep=False)].sort_values('uuid')
    display(duped.head(10))
```

## 4. Data Type Conversion

The KML parser reads all ExtendedData values as strings. Numeric and date columns need explicit conversion.

```python
# Numeric columns
for col in ['maxHeightAGL', 'topElevationAMSL', 'radius', 'kml_altitude']:
    if col in gdf.columns:
        gdf[col] = pd.to_numeric(gdf[col], errors='coerce')

# Boolean columns
for col in ['small']:
    if col in gdf.columns:
        gdf[col] = gdf[col].map({'true': True, 'false': False, 'True': True, 'False': False})

# Date columns
for col in ['effectiveDate', 'modified']:
    if col in gdf.columns:
        gdf[col] = pd.to_datetime(gdf[col], format='ISO8601', errors='coerce')

# Verify
gdf[['maxHeightAGL', 'topElevationAMSL', 'radius', 'effectiveDate']].dtypes
```

## 5. Descriptive Statistics: Numerical Features

```python
gdf[['maxHeightAGL', 'topElevationAMSL', 'radius', 'longitude', 'latitude']].describe().round(2)
```

## 6. Categorical Features

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

plot_cols = ['obstacleType', 'marking', 'lighting', 'group']
for ax, col in zip(axes.flatten(), plot_cols):
    counts = gdf[col].value_counts(dropna=False).head(10)
    counts.plot(kind='barh', ax=ax, color='steelblue', alpha=0.7)
    ax.set_title(f'{col}  ({gdf[col].nunique(dropna=False)} unique)')
    ax.invert_yaxis()

    for i, v in enumerate(counts.values):
        ax.text(v + 20, i, f'{v:,}', va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
airport_counts = gdf['airport'].value_counts(dropna=False).head(16)
airport_counts.name = 'count'
airport_df = airport_counts.to_frame()
airport_df['pct'] = (airport_df['count'] / len(gdf) * 100).round(1)
airport_df
```

## 7. Distributions: Numerical Features

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, col, title in zip(axes,
    ['maxHeightAGL', 'topElevationAMSL', 'radius'],
    ['Height Above Ground (m)', 'Elevation AMSL (m)', 'Radius (m)']):
    if col in gdf.columns:
        data = gdf[col].dropna()
        sns.histplot(data, kde=True, ax=ax, bins=50, alpha=0.4)
        ax.set_title(title)
        ax.set_xlabel(col)

        median = data.median()
        ax.axvline(median, color='red', linestyle='--', alpha=0.7, label=f'Median: {median:.0f}')
        ax.legend()

plt.tight_layout()
plt.show()
```

## 8. Outlier Assessment

```python
numeric_cols = ['maxHeightAGL', 'topElevationAMSL', 'radius']
numeric_cols = [c for c in numeric_cols if c in gdf.columns]

fig, axes = plt.subplots(1, len(numeric_cols), figsize=(6 * len(numeric_cols), 5))
if len(numeric_cols) == 1:
    axes = [axes]

for ax, col in zip(axes, numeric_cols):
    data = gdf[col].dropna()
    sns.boxplot(y=data, ax=ax)

    Q1, Q3 = data.quantile(0.25), data.quantile(0.75)
    IQR = Q3 - Q1
    n_outliers = ((data < Q1 - 1.5 * IQR) | (data > Q3 + 1.5 * IQR)).sum()
    pct = n_outliers / len(data) * 100

    ax.set_title(col)
    ax.text(0.5, -0.12, f'Outliers: {n_outliers} ({pct:.1f}%)',
            transform=ax.transAxes, ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```

## 9. Obstacle Type Distribution

```python
fig, ax = plt.subplots(figsize=(12, 6))

type_counts = gdf['obstacleType'].value_counts()
type_counts.plot(kind='barh', ax=ax, color='steelblue', alpha=0.7)

ax.set_xlabel('Count')
ax.set_ylabel('')
ax.set_title('Obstacle Count by Type')
ax.invert_yaxis()

for i, (count, name) in enumerate(zip(type_counts.values, type_counts.index)):
    ax.text(count + 50, i, f'{count:,}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

## 10. Geographic Distribution

```python
fig, ax = plt.subplots(figsize=(12, 10))

scatter = ax.scatter(
    gdf['longitude'],
    gdf['latitude'],
    c=gdf['maxHeightAGL'].clip(upper=200),
    cmap='YlOrRd',
    s=3,
    alpha=0.5
)

cbar = plt.colorbar(scatter, ax=ax, shrink=0.7)
cbar.set_label('Height AGL (m, clipped at 200)')

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title(f'Swiss Air Navigation Obstacles (n={len(gdf):,})')
ax.set_aspect('equal')

plt.tight_layout()
plt.show()
```

## 11. Correlation: Numerical Features

```python
numeric_for_corr = ['maxHeightAGL', 'topElevationAMSL', 'longitude', 'latitude']
numeric_for_corr = [c for c in numeric_for_corr if c in gdf.columns]

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(gdf[numeric_for_corr].corr(), annot=True, cmap='coolwarm', fmt='.2f', ax=ax)
ax.set_title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## Conclusion

The dataset comprises 14,334 registered air navigation obstacles across Switzerland, sourced from the
Federal Office of Civil Aviation (BAZL). All records carry a unique UUID with zero duplicates, and the
core features (obstacleType, maxHeightAGL, topElevationAMSL) are fully populated. Notable null rates
in airport (77.4%) and lighting (44.4%) are structurally expected, not data quality issues. The small
column is entirely null and will be dropped. Linear infrastructure (transmission lines, poles,
catenaries) accounts for 80% of all obstacles, which will require attention during clustering to avoid
masking subtler spatial patterns. All numerical features are right-skewed with outlier rates below 10%,
representing real structures (alpine cable cars, tall transmission towers) rather than measurement
errors. The geographic distribution reveals dense coverage across the Swiss Plateau, with visible gaps
in alpine terrain that represent the spatial signal the subsequent clustering analysis will target.
