```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import h3


sns.set_style('whitegrid')
```

## 1. Load Cleaned Dataset

```python
gdf = gpd.read_parquet('../data/processed/obstacles_clean.parquet')
print(f"Loaded: {len(gdf)} obstacles, {len(gdf.columns)} columns")
gdf.head(10)
```

## 2. Drop Unused Columns

The `small` column is entirely null (confirmed in Notebook 01) and `kml_altitude` is redundant with `topElevationAMSL`. The `source` column is constant ("main") since the small obstacles dataset was unavailable.

```python
drop_cols = [c for c in ['small', 'kml_altitude', 'source'] if c in gdf.columns]
gdf = gdf.drop(columns=drop_cols)
print(f"Dropped: {drop_cols}")
print(f"Remaining: {len(gdf.columns)} columns")
```

## 3. Assign Obstacles to H3 Hexagonal Grid

[H3](https://h3geo.org/) is Uber's hierarchical hexagonal grid system. Each resolution level divides the world into hexagonal cells of a specific size. We use **resolution 7**, which produces cells of approximately 5.16 km² — large enough to contain multiple obstacles per cell, small enough to capture regional variation across Switzerland.

| Resolution | Avg Cell Area | Avg Edge Length |
|:---:|:---:|:---:|
| 6 | 36.13 km² | 3.23 km |
| **7** | **5.16 km²** | **1.22 km** |
| 8 | 0.74 km² | 0.46 km |

```python
H3_RESOLUTION = 7

gdf['h3_index'] = gdf.apply(
    lambda row: h3.latlng_to_cell(row['latitude'], row['longitude'], H3_RESOLUTION),
    axis=1
)

n_cells = gdf['h3_index'].nunique()
print(f"Assigned {len(gdf)} obstacles to {n_cells} hexagonal cells")
print(f"Average obstacles per cell: {len(gdf) / n_cells:.1f}")
```

```python
obs_per_cell = gdf.groupby('h3_index').size()

fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(obs_per_cell, bins=50, kde=True, ax=ax, alpha=0.4)
ax.set_xlabel('Obstacles per Cell')
ax.set_ylabel('Count')
ax.set_title(f'Distribution of Obstacles per H3 Cell (n={n_cells} cells)')
ax.axvline(obs_per_cell.median(), color='red', linestyle='--',
           label=f'Median: {obs_per_cell.median():.0f}')
ax.legend()
plt.tight_layout()
plt.show()
```

## 4. Engineer Grid-Level Features

For each hexagonal cell, we compute the following features:

**Density:** total obstacle count in the cell.

**Height statistics:** mean, max, and standard deviation of obstacle height above ground (maxHeightAGL), plus mean terrain elevation (topElevationAMSL) as a proxy for whether the cell is in the lowlands or in alpine terrain.

**Type composition:** percentage of obstacles that are linear infrastructure (transmission lines, poles, catenaries) vs. percentage that are vertical structures (buildings, stacks, windmills). Also the number of distinct obstacle types present as a diversity measure.

**Risk indicators:** percentage of obstacles with lighting (relevant for night operations) and whether any obstacle in the cell is associated with an airport.

```python
LINEAR_TYPES = {'TRANSMISSION_LINE', 'POLE', 'CATENARY'}
VERTICAL_TYPES = {'BUILDING', 'STACK', 'WINDMILL', 'CRANE'}

def engineer_features(group):
    n = len(group)
    return pd.Series({
        # Density
        'obstacle_count': n,

        # Height statistics
        'height_mean': group['maxHeightAGL'].mean(),
        'height_max': group['maxHeightAGL'].max(),
        'height_std': group['maxHeightAGL'].std() if n > 1 else 0,

        # Terrain elevation
        'elevation_mean': group['topElevationAMSL'].mean(),

        # Type composition
        'pct_linear': (group['obstacleType'].isin(LINEAR_TYPES).sum() / n * 100),
        'pct_vertical': (group['obstacleType'].isin(VERTICAL_TYPES).sum() / n * 100),
        'pct_diversity': len(group['obstacleType'].unique()),

        # Risk indicators
        'pct_lighted': (group['lighting'].isin(['LIGHTED','LOW','MEDIUM', 'HIGH']).sum() / n * 100),
        'has_airport': int(group['airport'].notna().any()),

        # Cell centroid (for mapping later)
        'latitude': group['latitude'].mean(),
        'longitude': group['longitude'].mean(),
    })
grid = gdf.groupby('h3_index').apply(engineer_features).reset_index()
print(f"Grid features: {len(grid)} cells, {len(grid.columns)} columns")
grid.head()
```
