## 2. Drop Unused Columns

```python
drop_cols = [c for c in ['small', 'kml_altitude', 'source'] if c in gdf.columns]
gdf = gdf.drop(columns=drop_cols)
print(f"Dropped: {drop_cols}")
print(f"Remaining: {len(gdf.columns)} columns")
```
