# Corridor Profiling & Operational Assessment

**Objective:** Translate the K-Means k=4 clustering into actionable airspace risk profiles, identify low-obstacle corridors suitable for UAS transit, demonstrate scalability with a cantonal zoom at higher grid resolution, and produce final visualisations for the report.

**Input:** `data/processed/grid_clustered.parquet` (3,976 cells with cluster labels)
**Input:** `data/processed/obstacles_clean.parquet` (14,334 raw obstacles for cantonal zoom)

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import h3
import folium
import requests

```

```python
sns.set_style('whitegrid')
```

## Load Data

```python
grid = pd.read_parquet('../data/processed/grid_clustered.parquet')
gdf = gpd.read_parquet('../data/processed/obstacles_clean.parquet')

CLUSTER_NAMES = {
    0: 'Plateau Transmission',
    1: 'Airport & Urban',
    2: 'Alpine High-Rise',
    3: 'Urban Vertical'
}
COLORS = {0: '#4C72B0', 1: '#DD8452', 2: '#55A868', 3: '#C44E52'}

grid['cluster_name'] = grid['cluster_kmeans'].map(CLUSTER_NAMES)
print(f"Grid: {len(grid)} cells")
print(f"Obstacles: {len(gdf)} points")
```

## Operational Risk Profiles

Each cluster is characterised by its obstacle profile and assigned an operational risk assessment from a UAS corridor planning perspective.

```python
feature_cols = ['obstacle_count', 'height_mean', 'height_max', 'height_std',
                'elevation_mean', 'pct_linear', 'pct_vertical', 'pct_diversity',
                'pct_lighted', 'has_airport']

profiles = grid.groupby('cluster_kmeans')[feature_cols].agg(['mean', 'median']).round(1)
profiles.columns = ['_'.join(col) for col in profiles.columns]
profiles['count'] = grid.groupby('cluster_kmeans')['h3_index'].count()
profiles['share_pct'] = (profiles['count'] / len(grid) * 100).round(1)
profiles.index = profiles.index.map(CLUSTER_NAMES)
profiles
```

### Risk Classification

Based on the cluster profiles, each zone is assigned an operational risk level for low-altitude UAS operations:

| Zone | Risk Level | Rationale |
|:---|:---:|:---|
| **Plateau Transmission** | MODERATE | High linear obstacle density (91% transmission lines), but predictable geometry and low height variance. Obstacles follow known power grid corridors. |
| **Airport & Urban** | HIGH | Highest obstacle density, diverse obstacle types, airport proximity requires coordination with air traffic control. Restricted airspace likely. |
| **Alpine High-Rise** | HIGH | Extreme obstacle heights (mean 107m, max 173m) at high elevation. Cable car lines and valley crossings create unpredictable vertical hazards. Limited lighting. |
| **Urban Vertical** | MODERATE–HIGH | Dense vertical structures (82% buildings/stacks/cranes), but 45% lighted. Manageable with proper urban flight planning, but dynamic obstacles (cranes) add uncertainty. |

## Low-Obstacle Corridor Identification

Cells with low obstacle density, low height, and no airport proximity are candidates for low-risk UAS transit corridors. We define corridor cells as those meeting all of the following criteria:
- obstacle_count <= 2
- height_max <= 50m
- has_airport == 0
- pct_lighted == 0 (no special structures requiring lighting)

```python
corridor_mask = (
    (grid['obstacle_count'] <= 2) &
    (grid['height_max'] <= 50) &
    (grid['has_airport'] == 0) &
    (grid['pct_lighted'] == 0)
)

grid['is_corridor'] = corridor_mask.astype(int)
n_corridor = corridor_mask.sum()
print(f"Low-risk corridor cells: {n_corridor} / {len(grid)} ({n_corridor/len(grid)*100:.1f}%)")
print(f"\nCorridor cells by cluster:")
print(grid.loc[corridor_mask, 'cluster_name'].value_counts())
```

```python
fig, ax = plt.subplots(figsize=(16, 11))

# Background: all cells by cluster
for i, name in CLUSTER_NAMES.items():
    mask = grid['cluster_kmeans'] == i
    ax.scatter(grid.loc[mask, 'longitude'], grid.loc[mask, 'latitude'],
               c=COLORS[i], s=8, alpha=0.2, label=f'{name}')

# Highlight corridors
corridors = grid[corridor_mask]
ax.scatter(corridors['longitude'], corridors['latitude'],
           c='lime', s=20, alpha=0.8, edgecolors='darkgreen', linewidths=0.5,
           label=f'Low-Risk Corridor ({n_corridor} cells)', zorder=5)

ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)
ax.set_title('Swiss Airspace - Low-Risk UAS Transit Corridors', fontsize=14)
ax.set_aspect('equal')
ax.legend(fontsize=10, loc='lower left', framealpha=0.9)
plt.tight_layout()
plt.show()
```

```python
print("Corridor Cell Statistics:")
print(f"  Mean elevation: {corridors['elevation_mean'].mean():.0f}m AMSL")
print(f"  Elevation range: {corridors['elevation_mean'].min():.0f}m - {corridors['elevation_mean'].max():.0f}m")
print(f"  Mean obstacle count: {corridors['obstacle_count'].mean():.1f}")
print(f"  Mean height: {corridors['height_mean'].mean():.1f}m AGL")
print(f"\nGeographic extent:")
print(f"  Latitude:  {corridors['latitude'].min():.2f}° - {corridors['latitude'].max():.2f}°")
print(f"  Longitude: {corridors['longitude'].min():.2f}° - {corridors['longitude'].max():.2f}°")
```

### Interactive Corridor Map

```python


url = ("https://api3.geo.admin.ch/rest/services/api/MapServer/"
       "ch.swisstopo.swissboundaries3d-land-flaeche.fill/CH"
       "?geometryFormat=geojson&sr=4326")

response = requests.get(url)
feature = response.json()['feature']

# MultiPolygon: find the largest ring
all_rings = feature['geometry']['coordinates']
largest_ring = max(all_rings, key=lambda poly: len(poly[0]))[0]

# Convert to (lat, lon) tuples with explicit float
swiss_coords_wgs84 = [(float(c[1]), float(c[0])) for c in largest_ring]
print(f"Border polygon: {len(swiss_coords_wgs84)} points")

# Simplify for H3
step = max(1, len(swiss_coords_wgs84) // 500)
swiss_simplified = swiss_coords_wgs84[::step]
print(f"Simplified to {len(swiss_simplified)} points")

# Verify types
print(f"First 3 coords: {swiss_simplified[:3]}")
print(f"Type check: {type(swiss_simplified[0])}, {type(swiss_simplified[0][0])}")

swiss_poly = h3.LatLngPoly(swiss_simplified)
all_h3_cells = h3.polygon_to_cells(swiss_poly, 7)

occupied_cells = set(grid['h3_index'].values)
empty_cells = set(all_h3_cells) - occupied_cells
print(f"H3 cells within Switzerland: {len(all_h3_cells)}")
print(f"Occupied (obstacles): {len(occupied_cells)}")
print(f"Empty (clear airspace): {len(empty_cells)}")

def h3_to_polygon(h3_index):
    boundaries = h3.cell_to_boundary(h3_index)
    return [(lat, lng) for lat, lng in boundaries]

RISK_COLORS = {
    'Plateau Transmission': '#4C72B0',   # blue
    'Airport & Urban': '#CC3333',         # red
    'Alpine High-Rise': '#DD8452',        # orange
    'Urban Vertical': '#8855AA'           # violet
}
CLEAR_COLOR = '#44AA44'  # green

m = folium.Map(location=[46.8, 8.2], zoom_start=8, tiles='CartoDB positron')

# Draw empty cells first (green)
for cell_id in empty_cells:
    polygon = h3_to_polygon(cell_id)
    folium.Polygon(
        locations=polygon,
        color=CLEAR_COLOR,
        fill=True,
        fill_color=CLEAR_COLOR,
        fill_opacity=0.4,
        weight=0.5,
        popup='Clear Airspace',
    ).add_to(m)

# Draw occupied cells on top
for _, row in grid.iterrows():
    name = row['cluster_name']
    color = RISK_COLORS[name]
    polygon = h3_to_polygon(row['h3_index'])

    popup_text = (
        f"<b>{name}</b><br>"
        f"Obstacles: {row['obstacle_count']:.0f}<br>"
        f"Height mean: {row['height_mean']:.0f}m<br>"
        f"Elevation: {row['elevation_mean']:.0f}m AMSL"
    )

    folium.Polygon(
        locations=polygon,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6,
        weight=1,
        popup=folium.Popup(popup_text, max_width=200),
    ).add_to(m)


legend_html = """
<div style="position: fixed; bottom: 30px; left: 30px; z-index: 1000;
     background: white; padding: 15px; border-radius: 5px;
     border: 2px solid grey; font-size: 13px; line-height: 1.8;">
<b>Airspace Risk Zones</b><br>
<span style="color: #44AA44;">&#9632;</span> Clear Airspace (no obstacles)<br>
<span style="color: #4C72B0;">&#9632;</span> Plateau Transmission (low linear)<br>
<span style="color: #8855AA;">&#9632;</span> Urban Vertical (buildings/industry)<br>
<span style="color: #DD8452;">&#9632;</span> Alpine High-Rise (cable cars/towers)<br>
<span style="color: #CC3333;">&#9632;</span> Airport &amp; Urban (high density)
</div>
"""
m.get_root().html.add_child(folium.Element(legend_html))

m.save('../reports/figures/risk_zone_map.html')
print("Risk zone map saved")
m
```

## Cantonal Zoom - Higher Resolution Analysis

To demonstrate scalability, we re-run the grid aggregation at H3 resolution 9 (~0.1 km²) for a single canton. This provides the granularity needed for operational route planning, as opposed to the strategic overview at resolution 7.

We select **Canton Zürich** (approximate bounding box) as it contains the highest-density airspace zone (LSZH) alongside suburban and rural areas.

```python
# Canton Zurich approximate bounds
ZH_LAT_MIN, ZH_LAT_MAX = 47.15, 47.70
ZH_LON_MIN, ZH_LON_MAX = 8.35, 8.99

# Filter obstacles within Zürich
zh_mask = (
    (gdf['latitude'] >= ZH_LAT_MIN) & (gdf['latitude'] <= ZH_LAT_MAX) &
    (gdf['longitude'] >= ZH_LON_MIN) & (gdf['longitude'] <= ZH_LON_MAX)
)
gdf_zh = gdf[zh_mask].copy()
print(f"Obstacles in Canton Zürich: {len(gdf_zh)}")

# Assign H3 resolution 9
H3_RES_ZOOM = 9
gdf_zh['h3_zoom'] = gdf_zh.apply(
    lambda row: h3.latlng_to_cell(row['latitude'], row['longitude'], H3_RES_ZOOM),
    axis=1
)
print(f"H3 cells at resolution {H3_RES_ZOOM}: {gdf_zh['h3_zoom'].nunique()}")
```

```python
LINEAR_TYPES = {'TRANSMISSION_LINE', 'POLE', 'CATENARY'}
VERTICAL_TYPES = {'BUILDING', 'STACK', 'WINDMILL', 'CRANE'}

def engineer_features(group):
    n = len(group)
    return pd.Series({
        'obstacle_count': n,
        'height_mean': group['maxHeightAGL'].mean(),
        'height_max': group['maxHeightAGL'].max(),
        'height_std': group['maxHeightAGL'].std() if n > 1 else 0.0,
        'elevation_mean': group['topElevationAMSL'].mean(),
        'pct_linear': (group['obstacleType'].isin(LINEAR_TYPES).sum() / n * 100),
        'pct_vertical': (group['obstacleType'].isin(VERTICAL_TYPES).sum() / n * 100),
        'pct_diversity': group['obstacleType'].nunique(),
        'pct_lighted': (group['lighting'].isin(['LIGHTED', 'LOW', 'MEDIUM', 'HIGH']).sum() / n * 100),
        'has_airport': int(group['airport'].notna().any()),
        'latitude': group['latitude'].mean(),
        'longitude': group['longitude'].mean(),
    })

grid_zh = gdf_zh.groupby('h3_zoom').apply(engineer_features).reset_index()
print(f"Zürich grid: {len(grid_zh)} cells at resolution {H3_RES_ZOOM}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

feature_cols_zoom = ['obstacle_count', 'height_mean', 'height_max', 'height_std',
                     'elevation_mean', 'pct_linear', 'pct_vertical', 'pct_diversity',
                     'pct_lighted', 'has_airport']

X_zh = grid_zh[feature_cols_zoom].values
scaler_zh = StandardScaler()
X_zh_scaled = scaler_zh.fit_transform(X_zh)

km_zh = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
grid_zh['cluster'] = km_zh.fit_predict(X_zh_scaled)

# Map to names based on profile similarity
# Check which cluster has highest pct_linear, highest has_airport, etc.
zh_profiles = grid_zh.groupby('cluster')[feature_cols_zoom].mean()
zh_profiles = grid_zh.groupby('cluster')[feature_cols_zoom].mean().round(1)
zh_profiles['count'] = grid_zh.groupby('cluster')['h3_zoom'].count()
zh_profiles['share_pct'] = (zh_profiles['count'] / len(grid_zh) * 100).round(1)
zh_profiles
```

```python
# Identify clusters by dominant feature
zh_names = {}
for c in range(4):
    p = zh_profiles.loc[c]
    if p['has_airport'] > 0.5:
        zh_names[c] = 'Airport & Urban'
    elif p['pct_vertical'] > 50:
        zh_names[c] = 'Urban Vertical'
    elif p['elevation_mean'] > 800 and p['height_mean'] > 60:
        zh_names[c] = 'Alpine High-Rise'
    else:
        zh_names[c] = 'Plateau Transmission'

grid_zh['cluster_name'] = grid_zh['cluster'].map(zh_names)
print("Cluster name mapping:")
for k, v in zh_names.items():
    n = (grid_zh['cluster'] == k).sum()
    print(f"  Cluster {k} → {v} ({n} cells)")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22, 9))

# Left: Resolution 7 (Switzerland-wide, cropped to ZH)
zh_grid_r7 = grid[
    (grid['latitude'] >= ZH_LAT_MIN) & (grid['latitude'] <= ZH_LAT_MAX) &
    (grid['longitude'] >= ZH_LON_MIN) & (grid['longitude'] <= ZH_LON_MAX)
]

for i, name in CLUSTER_NAMES.items():
    mask = zh_grid_r7['cluster_kmeans'] == i
    if mask.any():
        ax1.scatter(zh_grid_r7.loc[mask, 'longitude'], zh_grid_r7.loc[mask, 'latitude'],
                    c=COLORS[i], s=40, alpha=0.6, label=name)

ax1.set_title(f'Canton Zürich - Resolution 7 ({len(zh_grid_r7)} cells, ~5 km²)', fontsize=13)
ax1.set_xlabel('Longitude')
ax1.set_ylabel('Latitude')
ax1.set_aspect('equal')
ax1.legend(fontsize=9)

# Right: Resolution 9
color_map_zh = {}
for c, name in zh_names.items():
    for orig_id, orig_name in CLUSTER_NAMES.items():
        if name == orig_name:
            color_map_zh[c] = COLORS[orig_id]
            break

for c, name in zh_names.items():
    mask = grid_zh['cluster'] == c
    ax2.scatter(grid_zh.loc[mask, 'longitude'], grid_zh.loc[mask, 'latitude'],
                c=color_map_zh.get(c, 'gray'), s=8, alpha=0.6, label=name)

ax2.set_title(f'Canton Zürich - Resolution 9 ({len(grid_zh)} cells, ~0.1 km²)', fontsize=13)
ax2.set_xlabel('Longitude')
ax2.set_ylabel('Latitude')
ax2.set_aspect('equal')

# Deduplicate legend
handles, labels = ax2.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax2.legend(by_label.values(), by_label.keys(), fontsize=9)

plt.tight_layout()
plt.show()
```

```python
url_zh = ("https://api3.geo.admin.ch/rest/services/api/MapServer/"
          "ch.swisstopo.swissboundaries3d-kanton-flaeche.fill/1"
          "?geometryFormat=geojson&sr=4326")

response_zh = requests.get(url_zh)
feature_zh = response_zh.json()['feature']
print(f"Canton: {feature_zh['properties']}")

# Extract largest ring from MultiPolygon
all_rings_zh = feature_zh['geometry']['coordinates']
largest_ring_zh = max(all_rings_zh, key=lambda poly: len(poly[0]))[0]

zh_coords = [(float(c[1]), float(c[0])) for c in largest_ring_zh]
print(f"Zürich border: {len(zh_coords)} points")

# Simplify
step_zh = max(1, len(zh_coords) // 300)
zh_simplified = zh_coords[::step_zh]
print(f"Simplified to {len(zh_simplified)} points")

zh_poly = h3.LatLngPoly(zh_simplified)
zh_h3_cells = h3.polygon_to_cells(zh_poly, H3_RES_ZOOM)

zh_occupied = set(grid_zh['h3_zoom'].values)
zh_empty = set(zh_h3_cells) - zh_occupied
print(f"Zürich H3 cells: {len(zh_h3_cells)} total, {len(zh_occupied)} occupied, {len(zh_empty)} clear")

RISK_COLORS_ZH = {
    'Plateau Transmission': '#4C72B0',
    'Airport & Urban': '#CC3333',
    'Alpine High-Rise': '#DD8452',
    'Urban Vertical': '#8855AA'
}
CLEAR_COLOR = '#44AA44'

m_zh = folium.Map(location=[47.38, 8.55], zoom_start=11, tiles='CartoDB positron')

# Empty cells (green)
for cell_id in zh_empty:
    polygon = h3_to_polygon(cell_id)
    folium.Polygon(
        locations=polygon,
        color=CLEAR_COLOR,
        fill=True,
        fill_color=CLEAR_COLOR,
        fill_opacity=0.4,
        weight=0.5,
        popup='Clear Airspace',
    ).add_to(m_zh)

# Occupied cells
for _, row in grid_zh.iterrows():
    c = int(row['cluster'])
    name = zh_names[c]
    color = RISK_COLORS_ZH.get(name, 'gray')
    polygon = h3_to_polygon(row['h3_zoom'])

    popup_text = (
        f"<b>{name}</b><br>"
        f"Obstacles: {row['obstacle_count']:.0f}<br>"
        f"Height mean: {row['height_mean']:.0f}m<br>"
        f"Max height: {row['height_max']:.0f}m<br>"
        f"Elevation: {row['elevation_mean']:.0f}m<br>"
        f"Linear: {row['pct_linear']:.0f}%<br>"
        f"Vertical: {row['pct_vertical']:.0f}%"
    )

    folium.Polygon(
        locations=polygon,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6,
        weight=1,
        popup=folium.Popup(popup_text, max_width=200),
    ).add_to(m_zh)

legend_html = """
<div style="position: fixed; bottom: 30px; left: 30px; z-index: 1000;
     background: white; padding: 15px; border-radius: 5px;
     border: 2px solid grey; font-size: 13px; line-height: 1.8;">
<b>Zürich Airspace (Resolution 9)</b><br>
<span style="color: #44AA44;">&#9632;</span> Clear Airspace<br>
<span style="color: #4C72B0;">&#9632;</span> Plateau Transmission<br>
<span style="color: #8855AA;">&#9632;</span> Urban Vertical<br>
<span style="color: #DD8452;">&#9632;</span> Alpine High-Rise<br>
<span style="color: #CC3333;">&#9632;</span> Airport &amp; Urban
</div>
"""
m_zh.get_root().html.add_child(folium.Element(legend_html))

map_path_zh = '../reports/figures/zurich_zoom_map.html'
m_zh.save(map_path_zh)
print(f"Interactive Zürich map saved to {map_path_zh}")
m_zh
```
