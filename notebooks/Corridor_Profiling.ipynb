# Corridor Profiling & Operational Assessment

**Objective:** Translate the K-Means k=4 clustering into actionable airspace risk profiles, identify low-obstacle corridors suitable for UAS transit, demonstrate scalability with a cantonal zoom at higher grid resolution, and produce final visualisations for the report.

**Input:** `data/processed/grid_clustered.parquet` (3,976 cells with cluster labels)
**Input:** `data/processed/obstacles_clean.parquet` (14,334 raw obstacles for cantonal zoom)

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import h3
import folium
from folium.plugins import HeatMap
```

```python
sns.set_style('whitegrid')
```

## Load Data

```python
grid = pd.read_parquet('../data/processed/grid_clustered.parquet')
gdf = gpd.read_parquet('../data/processed/obstacles_clean.parquet')

CLUSTER_NAMES = {
    0: 'Plateau Transmission',
    1: 'Airport & Urban',
    2: 'Alpine High-Rise',
    3: 'Urban Vertical'
}
COLORS = {0: '#4C72B0', 1: '#DD8452', 2: '#55A868', 3: '#C44E52'}

grid['cluster_name'] = grid['cluster_kmeans'].map(CLUSTER_NAMES)
print(f"Grid: {len(grid)} cells")
print(f"Obstacles: {len(gdf)} points")
```

## Operational Risk Profiles

Each cluster is characterised by its obstacle profile and assigned an operational risk assessment from a UAS corridor planning perspective.

```python
feature_cols = ['obstacle_count', 'height_mean', 'height_max', 'height_std',
                'elevation_mean', 'pct_linear', 'pct_vertical', 'pct_diversity',
                'pct_lighted', 'has_airport']

profiles = grid.groupby('cluster_kmeans')[feature_cols].agg(['mean', 'median']).round(1)
profiles.columns = ['_'.join(col) for col in profiles.columns]
profiles['count'] = grid.groupby('cluster_kmeans')['h3_index'].count()
profiles['share_pct'] = (profiles['count'] / len(grid) * 100).round(1)
profiles.index = profiles.index.map(CLUSTER_NAMES)
profiles
```
