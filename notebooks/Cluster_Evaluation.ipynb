```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm

try:
    import hdbscan
except ImportError:
    print("Installing hdbscan...")
    import subprocess
    subprocess.check_call(['pip', 'install', 'hdbscan'])
    import hdbscan

sns.set_style('whitegrid')
```

## 1. Load Grid Features

```python
grid = pd.read_parquet('../data/processed/grid_features.parquet')
print(f"Loaded: {len(grid)} grid cells")

feature_cols = ['obstacle_count', 'height_mean', 'height_max', 'height_std',
                'elevation_mean', 'pct_linear', 'pct_vertical', 'pct_diversity',
                'pct_lighted', 'has_airport']

X = grid[feature_cols].copy()
X.head()
```

## 2. Feature Scaling

K-Means and DBSCAN use Euclidean distance, which is dominated by features with large magnitudes (e.g. elevation_mean in the hundreds vs. has_airport as 0/1). StandardScaler normalises all features to zero mean and unit variance so that each feature contributes equally to distance calculations. HDBSCAN also benefits from scaling for the same reason.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Verify: mean ≈ 0, std ≈ 1
pd.DataFrame(X_scaled, columns=feature_cols).describe().round(2).loc[['mean', 'std']]
```

## 3. HDBSCAN (Primary Method)

HDBSCAN is the primary clustering method for this project because it does not require a pre-specified number of clusters, it can identify clusters of varying density (critical for a dataset where urban centres are dense and alpine regions are sparse), and it labels ambiguous points as noise (-1) rather than forcing them into a cluster. The key parameter `min_cluster_size` sets the smallest group of points that should be considered a cluster.

```python
# Test multiple min_cluster_size values
results_hdb = []

for mcs in [15, 25, 50, 75, 100]:
    clusterer = hdbscan.HDBSCAN(min_cluster_size=mcs, min_samples=5,
                                 metric='euclidean')
    labels = clusterer.fit_predict(X_scaled)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = (labels == -1).sum()
    noise_pct = n_noise / len(labels) * 100

    # Silhouette only on non-noise points
    if n_clusters >= 2:
        mask = labels != -1
        sil = silhouette_score(X_scaled[mask], labels[mask])
    else:
        sil = -1

    results_hdb.append({
        'min_cluster_size': mcs,
        'n_clusters': n_clusters,
        'noise': n_noise,
        'noise_pct': round(noise_pct, 1),
        'silhouette': round(sil, 3)
    })

results_hdb_df = pd.DataFrame(results_hdb)
results_hdb_df
```

### 3.1 Select HDBSCAN Configuration

We select the configuration that balances a reasonable number of clusters, acceptable noise ratio, and the highest silhouette score.

```python
# Select best based on results above, adjust min_cluster_size if needed
BEST_MCS = results_hdb_df.loc[results_hdb_df['silhouette'].idxmax(), 'min_cluster_size']
print(f"Selected min_cluster_size: {BEST_MCS}")

hdb = hdbscan.HDBSCAN(min_cluster_size=int(BEST_MCS), min_samples=5, metric='euclidean')
grid['cluster_hdbscan'] = hdb.fit_predict(X_scaled)

n_clusters = grid['cluster_hdbscan'].nunique() - (1 if -1 in grid['cluster_hdbscan'].values else 0)
n_noise = (grid['cluster_hdbscan'] == -1).sum()
print(f"Clusters: {n_clusters}, Noise: {n_noise} ({n_noise/len(grid)*100:.1f}%)")
```

```python
fig, ax = plt.subplots(figsize=(14, 10))

noise = grid[grid['cluster_hdbscan'] == -1]
clustered = grid[grid['cluster_hdbscan'] != -1]

ax.scatter(noise['longitude'], noise['latitude'], c='lightgray', s=8, alpha=0.3, label='Noise')
scatter = ax.scatter(clustered['longitude'], clustered['latitude'],
                     c=clustered['cluster_hdbscan'], cmap='tab10', s=12, alpha=0.6)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title(f'HDBSCAN Clusters (k={n_clusters}, noise={n_noise})')
ax.set_aspect('equal')
ax.legend()
plt.colorbar(scatter, ax=ax, label='Cluster', shrink=0.7)
plt.tight_layout()
plt.show()
```
