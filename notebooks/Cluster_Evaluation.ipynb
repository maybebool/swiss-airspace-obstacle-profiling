```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm

try:
    import hdbscan
except ImportError:
    print("Installing hdbscan...")
    import subprocess
    subprocess.check_call(['pip', 'install', 'hdbscan'])
    import hdbscan

sns.set_style('whitegrid')
```

## 1. Load Grid Features

```python
grid = pd.read_parquet('../data/processed/grid_features.parquet')
print(f"Loaded: {len(grid)} grid cells")

feature_cols = ['obstacle_count', 'height_mean', 'height_max', 'height_std',
                'elevation_mean', 'pct_linear', 'pct_vertical', 'pct_diversity',
                'pct_lighted', 'has_airport']

X = grid[feature_cols].copy()
X.head()
```

## 2. Feature Scaling

K-Means and DBSCAN use Euclidean distance, which is dominated by features with large magnitudes (e.g. elevation_mean in the hundreds vs. has_airport as 0/1). StandardScaler normalises all features to zero mean and unit variance so that each feature contributes equally to distance calculations. HDBSCAN also benefits from scaling for the same reason.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Verify: mean ≈ 0, std ≈ 1
pd.DataFrame(X_scaled, columns=feature_cols).describe().round(2).loc[['mean', 'std']]
```

## 3. HDBSCAN (Primary Method)

HDBSCAN is the primary clustering method for this project because it does not require a pre-specified number of clusters, it can identify clusters of varying density (critical for a dataset where urban centres are dense and alpine regions are sparse), and it labels ambiguous points as noise (-1) rather than forcing them into a cluster. The key parameter `min_cluster_size` sets the smallest group of points that should be considered a cluster.

```python
# Test multiple min_cluster_size values
results_hdb = []

for mcs in [15, 25, 50, 75, 100]:
    clusterer = hdbscan.HDBSCAN(min_cluster_size=mcs, min_samples=5,
                                 metric='euclidean')
    labels = clusterer.fit_predict(X_scaled)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = (labels == -1).sum()
    noise_pct = n_noise / len(labels) * 100

    # Silhouette only on non-noise points
    if n_clusters >= 2:
        mask = labels != -1
        sil = silhouette_score(X_scaled[mask], labels[mask])
    else:
        sil = -1

    results_hdb.append({
        'min_cluster_size': mcs,
        'n_clusters': n_clusters,
        'noise': n_noise,
        'noise_pct': round(noise_pct, 1),
        'silhouette': round(sil, 3)
    })

results_hdb_df = pd.DataFrame(results_hdb)
results_hdb_df
```

### 3.1 Select HDBSCAN Configuration

We select the configuration that balances a reasonable number of clusters, acceptable noise ratio, and the highest silhouette score.

```python
# Select best based on results above, adjust min_cluster_size if needed
BEST_MCS = results_hdb_df.loc[results_hdb_df['silhouette'].idxmax(), 'min_cluster_size']
print(f"Selected min_cluster_size: {BEST_MCS}")

hdb = hdbscan.HDBSCAN(min_cluster_size=int(BEST_MCS), min_samples=5, metric='euclidean')
grid['cluster_hdbscan'] = hdb.fit_predict(X_scaled)

n_clusters = grid['cluster_hdbscan'].nunique() - (1 if -1 in grid['cluster_hdbscan'].values else 0)
n_noise = (grid['cluster_hdbscan'] == -1).sum()
print(f"Clusters: {n_clusters}, Noise: {n_noise} ({n_noise/len(grid)*100:.1f}%)")
```

```python
fig, ax = plt.subplots(figsize=(14, 10))

noise = grid[grid['cluster_hdbscan'] == -1]
clustered = grid[grid['cluster_hdbscan'] != -1]

ax.scatter(noise['longitude'], noise['latitude'], c='lightgray', s=8, alpha=0.3, label='Noise')
scatter = ax.scatter(clustered['longitude'], clustered['latitude'],
                     c=clustered['cluster_hdbscan'], cmap='tab10', s=12, alpha=0.6)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title(f'HDBSCAN Clusters (k={n_clusters}, noise={n_noise})')
ax.set_aspect('equal')
ax.legend()
plt.colorbar(scatter, ax=ax, label='Cluster', shrink=0.7)
plt.tight_layout()
plt.show()
```

## 4. K-Means (Baseline Comparison)

K-Means serves as a baseline. Unlike HDBSCAN, it requires a pre-specified k and assigns every point to a cluster (no noise concept). We use the Elbow method and Silhouette analysis to find the optimal k.

```python
wcss = []
sil_scores = []
K = range(2, 11)

for k in K:
    km = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    wcss.append(km.inertia_)
    sil_scores.append(silhouette_score(X_scaled, labels))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(K, wcss, 'bo-')
ax1.set_xlabel('k')
ax1.set_ylabel('WCSS')
ax1.set_title('Elbow Method')

ax2.plot(K, sil_scores, 'ro-')
ax2.set_xlabel('k')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score by k')

plt.tight_layout()
plt.show()

for k, w, s in zip(K, wcss, sil_scores):
    print(f"k={k}: WCSS={w:.0f}, Silhouette={s:.3f}")
```

```python
# Select best k based on elbow + silhouette, adjust if needed
BEST_K = sil_scores.index(max(sil_scores)) + 2  # +2 because K starts at 2
print(f"Best k by silhouette: {BEST_K}")

km_best = KMeans(n_clusters=BEST_K, init='k-means++', random_state=42, n_init=10)
grid['cluster_kmeans'] = km_best.fit_predict(X_scaled)

print(f"Silhouette: {silhouette_score(X_scaled, grid['cluster_kmeans']):.3f}")
```

```python
fig, ax = plt.subplots(figsize=(14, 10))

scatter = ax.scatter(grid['longitude'], grid['latitude'],
                     c=grid['cluster_kmeans'], cmap='tab10', s=12, alpha=0.6)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title(f'K-Means Clusters (k={BEST_K})')
ax.set_aspect('equal')
plt.colorbar(scatter, ax=ax, label='Cluster', shrink=0.7)
plt.tight_layout()
plt.show()
```

## 5. DBSCAN (Density-Based Comparison)

DBSCAN is density-based like HDBSCAN but requires manual tuning of epsilon (neighbourhood radius). We test multiple epsilon values to find a reasonable configuration.

```python
results_db = []

for eps in [0.8, 1.0, 1.2, 1.5, 2.0, 2.5]:
    db = DBSCAN(eps=eps, min_samples=5)
    labels = db.fit_predict(X_scaled)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = (labels == -1).sum()
    noise_pct = n_noise / len(labels) * 100

    if n_clusters >= 2:
        mask = labels != -1
        if mask.sum() > n_clusters:
            sil = silhouette_score(X_scaled[mask], labels[mask])
        else:
            sil = -1
    else:
        sil = -1

    results_db.append({
        'eps': eps,
        'n_clusters': n_clusters,
        'noise': n_noise,
        'noise_pct': round(noise_pct, 1),
        'silhouette': round(sil, 3)
    })

results_db_df = pd.DataFrame(results_db)
results_db_df
```

```python
# Select best eps — adjust based on results above
valid = results_db_df[results_db_df['silhouette'] > 0]
if len(valid) > 0:
    BEST_EPS = valid.loc[valid['silhouette'].idxmax(), 'eps']
else:
    BEST_EPS = 1.5

print(f"Selected eps: {BEST_EPS}")

db_best = DBSCAN(eps=BEST_EPS, min_samples=5)
grid['cluster_dbscan'] = db_best.fit_predict(X_scaled)

n_clusters_db = len(set(grid['cluster_dbscan'])) - (1 if -1 in grid['cluster_dbscan'].values else 0)
n_noise_db = (grid['cluster_dbscan'] == -1).sum()
print(f"Clusters: {n_clusters_db}, Noise: {n_noise_db} ({n_noise_db/len(grid)*100:.1f}%)")
```

```python
fig, ax = plt.subplots(figsize=(14, 10))

noise = grid[grid['cluster_dbscan'] == -1]
clustered = grid[grid['cluster_dbscan'] != -1]

ax.scatter(noise['longitude'], noise['latitude'], c='lightgray', s=8, alpha=0.3, label='Noise')
scatter = ax.scatter(clustered['longitude'], clustered['latitude'],
                     c=clustered['cluster_dbscan'], cmap='tab10', s=12, alpha=0.6)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title(f'DBSCAN Clusters (eps={BEST_EPS}, k={n_clusters_db}, noise={n_noise_db})')
ax.set_aspect('equal')
ax.legend()
plt.colorbar(scatter, ax=ax, label='Cluster', shrink=0.7)
plt.tight_layout()
plt.show()
```

## 6. Method Comparison

```python
# Side-by-side geographic comparison
fig, axes = plt.subplots(1, 3, figsize=(24, 8))

methods = [
    ('cluster_hdbscan', 'HDBSCAN'),
    ('cluster_kmeans', 'K-Means'),
    ('cluster_dbscan', 'DBSCAN')
]

for ax, (col, title) in zip(axes, methods):
    noise = grid[grid[col] == -1] if -1 in grid[col].values else pd.DataFrame()
    clustered = grid[grid[col] != -1]

    if len(noise) > 0:
        ax.scatter(noise['longitude'], noise['latitude'],
                   c='lightgray', s=5, alpha=0.3)

    ax.scatter(clustered['longitude'], clustered['latitude'],
               c=clustered[col], cmap='tab10', s=5, alpha=0.6)

    n_c = grid[col].nunique() - (1 if -1 in grid[col].values else 0)
    n_n = (grid[col] == -1).sum()

    mask = grid[col] != -1
    if mask.sum() > 0 and n_c >= 2:
        sil = silhouette_score(X_scaled[mask], grid.loc[mask, col])
        ax.set_title(f'{title}\nk={n_c}, noise={n_n}, sil={sil:.3f}')
    else:
        ax.set_title(f'{title}\nk={n_c}, noise={n_n}')

    ax.set_aspect('equal')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

plt.tight_layout()
plt.show()
```

## 7. Silhouette Analysis: Selected Method

Silhouette plot for the chosen clustering to assess how well-separated each cluster is. Values near 1 indicate well-separated clusters, values near 0 indicate overlapping clusters, and negative values indicate potential misassignment.

```python
# Silhouette plot for HDBSCAN (excluding noise)
mask = grid['cluster_hdbscan'] != -1
X_clust = X_scaled[mask]
labels_clust = grid.loc[mask, 'cluster_hdbscan'].values

if len(set(labels_clust)) >= 2:
    sil_avg = silhouette_score(X_clust, labels_clust)
    sil_vals = silhouette_samples(X_clust, labels_clust)

    n_clusters = len(set(labels_clust))
    fig, ax = plt.subplots(figsize=(10, max(7, n_clusters * 1.5)))

    y_lower = 10
    for i in sorted(set(labels_clust)):
        cluster_sil = sil_vals[labels_clust == i]
        cluster_sil.sort()

        y_upper = y_lower + len(cluster_sil)
        color = cm.tab10(i / max(n_clusters, 1))
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_sil,
                         facecolor=color, alpha=0.7)
        ax.text(-0.05, y_lower + 0.5 * len(cluster_sil), str(i), fontsize=11)
        y_lower = y_upper + 10

    ax.axvline(x=sil_avg, color='red', linestyle='--',
               label=f'Average: {sil_avg:.3f}')
    ax.set_xlabel('Silhouette Coefficient')
    ax.set_ylabel('Cluster')
    ax.set_title('HDBSCAN Silhouette Plot')
    ax.legend()
    plt.tight_layout()
    plt.show()
else:
    print("Not enough clusters for silhouette analysis.")
```

```python
sil_avg = silhouette_score(X_scaled, grid['cluster_kmeans'])
sil_vals = silhouette_samples(X_scaled, grid['cluster_kmeans'])

fig, ax = plt.subplots(figsize=(10, 7))
y_lower = 10
for i in range(BEST_K):
    cluster_sil = sil_vals[grid['cluster_kmeans'] == i]
    cluster_sil.sort()
    y_upper = y_lower + len(cluster_sil)
    color = cm.tab10(i / BEST_K)
    ax.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_sil,
                     facecolor=color, alpha=0.7)
    ax.text(-0.05, y_lower + 0.5 * len(cluster_sil), str(i), fontsize=11)
    y_lower = y_upper + 10

ax.axvline(x=sil_avg, color='red', linestyle='--', label=f'Average: {sil_avg:.3f}')
ax.set_xlabel('Silhouette Coefficient')
ax.set_ylabel('Cluster')
ax.set_title('K-Means Silhouette Plot (k=4)')
ax.legend()
plt.tight_layout()
plt.show()
```

## Conclusion

Three clustering methods were evaluated on the 3,976 grid cells: HDBSCAN (13 clusters, 23.4% noise,
silhouette 0.229), K-Means (k=4, silhouette 0.363), and DBSCAN (11 clusters but 77% in a single
dominant cluster, effectively unusable). K-Means with k=4 was selected as the primary result based on
the highest silhouette score and the most operationally interpretable segmentation. The four clusters
form distinct airspace profiles: Plateau Transmission Corridors (63.9%, low-density linear
infrastructure), Airport & Urban High-Density zones (13.9%, highest obstacle count, airport proximity),
Alpine High-Rise Infrastructure (15.1%, extreme obstacle heights at 1,481m mean elevation), and Urban
Vertical & Industrial zones (7.1%, 82% vertical structures, 45% lighted). Silhouette analysis shows
soft boundaries between clusters 1 and 2, consistent with the gradual geographic transitions inherent
in spatial obstacle data. The clustered grid is saved for dimensionality reduction and corridor
profiling in subsequent notebooks.
