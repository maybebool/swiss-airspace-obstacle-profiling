```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import h3
```

## Load Cleaned Dataset

```python
gdf = gpd.read_parquet('../data/processed/obstacles_clean.parquet')
print(f"Loaded: {len(gdf)} obstacles, {len(gdf.columns)} columns")
gdf.head(10)
```

## Drop Unused Columns

The `small` column is entirely null (confirmed in Notebook 01) and `kml_altitude` is redundant with `topElevationAMSL`. The `source` column is constant ("main") since the small obstacles dataset was unavailable.

```python
drop_cols = [c for c in ['small', 'kml_altitude', 'source'] if c in gdf.columns]
gdf = gdf.drop(columns=drop_cols)
print(f"Dropped: {drop_cols}")
print(f"Remaining: {len(gdf.columns)} columns")
```

## Assign Obstacles to H3 Hexagonal Grid

[H3](https://h3geo.org/) is Uber's hierarchical hexagonal grid system. Each resolution level divides the world into hexagonal cells of a specific size. We use **resolution 7**, which produces cells of approximately 5.16 km² - large enough to contain multiple obstacles per cell, small enough to capture regional variation across Switzerland.

| Resolution | Avg Cell Area | Avg Edge Length |
|:---:|:---:|:---:|
| 6 | 36.13 km² | 3.23 km |
| **7** | **5.16 km²** | **1.22 km** |
| 8 | 0.74 km² | 0.46 km |

```python
H3_RESOLUTION = 7

gdf['h3_index'] = gdf.apply(
    lambda row: h3.latlng_to_cell(row['latitude'], row['longitude'], H3_RESOLUTION),
    axis=1
)

n_cells = gdf['h3_index'].nunique()
print(f"Assigned {len(gdf)} obstacles to {n_cells} hexagonal cells")
print(f"Average obstacles per cell: {len(gdf) / n_cells:.1f}")
```

```python
sns.set_style('whitegrid')
obs_per_cell = gdf.groupby('h3_index').size()

fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(obs_per_cell, bins=50, kde=True, ax=ax, alpha=0.4)
ax.set_xlabel('Obstacles per Cell')
ax.set_ylabel('Count')
ax.set_title(f'Distribution of Obstacles per H3 Cell (n={n_cells} cells)')
ax.axvline(obs_per_cell.median(), color='red', linestyle='--',
           label=f'Median: {obs_per_cell.median():.0f}')
ax.legend()
plt.tight_layout()
plt.show()
```

## Engineer Grid-Level Features

For each hexagonal cell, we compute the following features:

**Density:** total obstacle count in the cell.

**Height statistics:** mean, max, and standard deviation of obstacle height above ground (maxHeightAGL), plus mean terrain elevation (topElevationAMSL) as a proxy for whether the cell is in the lowlands or in alpine terrain.

**Type composition:** percentage of obstacles that are linear infrastructure (transmission lines, poles, catenaries) vs. percentage that are vertical structures (buildings, stacks, windmills). Also the number of distinct obstacle types present as a diversity measure.

**Risk indicators:** percentage of obstacles with lighting (relevant for night operations) and whether any obstacle in the cell is associated with an airport.

```python
LINEAR_TYPES = {'TRANSMISSION_LINE', 'POLE', 'CATENARY'}
VERTICAL_TYPES = {'BUILDING', 'STACK', 'WINDMILL', 'CRANE'}

def engineer_features(group):
    n = len(group)
    return pd.Series({
        # Density
        'obstacle_count': n,

        # Height statistics
        'height_mean': group['maxHeightAGL'].mean(),
        'height_max': group['maxHeightAGL'].max(),
        'height_std': group['maxHeightAGL'].std() if n > 1 else 0,

        # Terrain elevation
        'elevation_mean': group['topElevationAMSL'].mean(),

        # Type composition
        'pct_linear': (group['obstacleType'].isin(LINEAR_TYPES).sum() / n * 100),
        'pct_vertical': (group['obstacleType'].isin(VERTICAL_TYPES).sum() / n * 100),
        'pct_diversity': len(group['obstacleType'].unique()),

        # Risk indicators
        'pct_lighted': (group['lighting'].isin(['LIGHTED','LOW','MEDIUM', 'HIGH']).sum() / n * 100),
        'has_airport': int(group['airport'].notna().any()),

        # Cell centroid (for mapping later)
        'latitude': group['latitude'].mean(),
        'longitude': group['longitude'].mean(),
    })
grid = gdf.groupby('h3_index').apply(engineer_features).reset_index()
print(f"Grid features: {len(grid)} cells, {len(grid.columns)} columns")
grid.head()
```

```python
print(grid.columns.tolist())
```

## Grid Feature Overview

```python
feature_cols = ['obstacle_count', 'height_mean', 'height_max', 'height_std',
                'elevation_mean', 'pct_linear', 'pct_vertical', 'pct_diversity',
                'pct_lighted', 'has_airport']

grid[feature_cols].describe().round(2)
```

## Feature Distributions

```python
fig, axes = plt.subplots(3, 4, figsize=(20, 12))
axes = axes.flatten()

for i, col in enumerate(feature_cols):
    ax = axes[i]
    sns.histplot(grid[col], bins=40, kde=True, ax=ax, alpha=0.4)
    ax.set_title(col)
    ax.set_xlabel('')

# Hide unused subplot
for j in range(len(feature_cols), len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

## Feature Correlations

```python
fig, ax = plt.subplots(figsize=(10, 8))
corr = grid[feature_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', ax=ax,
            vmin=-1, vmax=1, center=0)
ax.set_title('Feature Correlation Matrix (Grid Level)')
plt.tight_layout()
plt.show()
```

## Geographic Preview: Obstacle Density

```python
fig, ax = plt.subplots(figsize=(14, 10))

scatter = ax.scatter(
    grid['longitude'],
    grid['latitude'],
    c=grid['obstacle_count'],
    cmap='YlOrRd',
    s=15,
    alpha=0.7
)

cbar = plt.colorbar(scatter, ax=ax, shrink=0.7)
cbar.set_label('Obstacles per Cell')

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title(f'Obstacle Density per H3 Cell (n={len(grid)} cells)')
ax.set_aspect('equal')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 10))

scatter = ax.scatter(
    grid['longitude'],
    grid['latitude'],
    c=grid['elevation_mean'],
    cmap='terrain',
    s=15,
    alpha=0.7
)

cbar = plt.colorbar(scatter, ax=ax, shrink=0.7)
cbar.set_label('Mean Elevation AMSL (m)')

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title(f'Terrain Elevation by Grid Cell')
ax.set_aspect('equal')
plt.tight_layout()
plt.show()
```

## Save Grid Features

```python
out_path = '../data/processed/grid_features.parquet'
grid.to_parquet(out_path, index=False)
print(f"Saved: {len(grid)} grid cells, {len(grid.columns)} columns → {out_path}")
```

## Conclusion

The 14,334 point-level obstacles were aggregated onto a hexagonal H3 grid (resolution 7, ~5 km² per
cell), producing 3,976 grid cells with 10 engineered features each. The feature set captures obstacle
density, height statistics (mean, max, standard deviation), terrain elevation, type composition
(percentage linear vs. vertical infrastructure), lighting coverage, and airport proximity. The
correlation matrix confirms that pct_linear and pct_vertical are inversely related (r=-0.66),
indicating cells are predominantly either transmission corridors or built-up zones. Geographic
visualisation shows dense, low-elevation coverage across the Swiss Plateau transitioning to sparse,
high-elevation cells in the Alps, with a few high-density hotspots at urban centres. This grid-level
dataset is the direct input for spatial clustering in Notebook 04.
