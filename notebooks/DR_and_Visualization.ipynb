# Dimensionality Reduction & Visualization

**Objective:** Verify and visualise the K-Means k=4 clustering through dimensionality reduction (PCA, t-SNE) and geographic overlay maps. Confirm that clusters correspond to operationally meaningful airspace zones.

**Input:** `data/processed/grid_clustered.parquet` (3,976 cells with cluster labels from Notebook 03)
**Output:** Visualisations and maps for the final report. No new parquet, this notebook is analysis-only.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import folium
from folium.plugins import MarkerCluster
```

## Load Clustered Grid

```python
grid = pd.read_parquet('../data/processed/grid_clustered.parquet')
print(f"Loaded: {len(grid)} grid cells")

feature_cols = ['obstacle_count', 'height_mean', 'height_max', 'height_std',
                'elevation_mean', 'pct_linear', 'pct_vertical', 'pct_diversity',
                'pct_lighted', 'has_airport']

X = grid[feature_cols].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Cluster labels and names for plotting
cluster_labels = grid['cluster_kmeans'].values
CLUSTER_NAMES = {
    0: 'Plateau Transmission',
    1: 'Airport & Urban',
    2: 'Alpine High-Rise',
    3: 'Urban Vertical'
}
grid['cluster_name'] = grid['cluster_kmeans'].map(CLUSTER_NAMES)
print(grid['cluster_name'].value_counts())
```

## PCA: Variance Decomposition

PCA reveals how much of the total variance in the 10 features can be explained by fewer components, and which features drive each component.

```python
sns.set_style('whitegrid')
pca_full = PCA(random_state=42)
pca_full.fit(X_scaled)

# Explained variance
cumvar = np.cumsum(pca_full.explained_variance_ratio_) * 100

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.bar(range(1, 11), pca_full.explained_variance_ratio_ * 100, alpha=0.6, color='steelblue')
ax1.plot(range(1, 11), cumvar, 'ro-')
ax1.set_xlabel('Principal Component')
ax1.set_ylabel('Variance Explained (%)')
ax1.set_title('PCA Variance Explained')
ax1.set_xticks(range(1, 11))

for i, (var, cum) in enumerate(zip(pca_full.explained_variance_ratio_ * 100, cumvar)):
    ax1.text(i + 1, cum + 1.5, f'{cum:.0f}%', ha='center', fontsize=9)

# How many components for 90%?
n_90 = np.argmax(cumvar >= 90) + 1
ax1.axhline(y=90, color='gray', linestyle='--', alpha=0.5, label=f'90% at PC{n_90}')
ax1.legend()

# Feature loadings for PC1 and PC2
loadings = pd.DataFrame(
    pca_full.components_[:2].T,
    columns=['PC1', 'PC2'],
    index=feature_cols
)
loadings.plot(kind='barh', ax=ax2, alpha=0.7)
ax2.set_title('Feature Loadings (PC1 & PC2)')
ax2.set_xlabel('Loading')
ax2.axvline(x=0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()

print(f"\nPC1 explains {pca_full.explained_variance_ratio_[0]*100:.1f}% of variance")
print(f"PC2 explains {pca_full.explained_variance_ratio_[1]*100:.1f}% of variance")
print(f"Components needed for 90% variance: {n_90}")
```

### PCA Projection: Clusters in PC1/PC2 Space

```python
pca_2d = PCA(n_components=2, random_state=42)
pca_result = pca_2d.fit_transform(X_scaled)

fig, ax = plt.subplots(figsize=(10, 8))

colors = ['#4C72B0', '#ad0e0e', '#55A868', '#a19d2f']
for i, name in CLUSTER_NAMES.items():
    mask = cluster_labels == i
    ax.scatter(pca_result[mask, 0], pca_result[mask, 1],
               c=colors[i], s=15, alpha=0.5, label=f'{name} (n={mask.sum()})')

ax.set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}%)')
ax.set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}%)')
ax.set_title('K-Means Clusters in PCA Space')
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

### PCA Biplot: Feature Vectors

The biplot overlays the feature loading vectors onto the PCA scatter, showing which features drive the separation between clusters.

```python
fig, ax = plt.subplots(figsize=(12, 9))

# Scatter
for i, name in CLUSTER_NAMES.items():
    mask = cluster_labels == i
    ax.scatter(pca_result[mask, 0], pca_result[mask, 1],
               c=colors[i], s=10, alpha=0.3, label=f'{name}')

# Loading vectors
loadings_2d = pca_2d.components_.T
scale = 4  # scale arrows for visibility

for j, feat in enumerate(feature_cols):
    ax.arrow(0, 0, loadings_2d[j, 0] * scale, loadings_2d[j, 1] * scale,
             head_width=0.08, head_length=0.05, fc='black', ec='black', alpha=0.7)
    ax.text(loadings_2d[j, 0] * scale * 1.15, loadings_2d[j, 1] * scale * 1.15,
            feat, fontsize=9, ha='center', va='center',
            bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))

ax.set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}%)')
ax.set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}%)')
ax.set_title('PCA Biplot — Feature Directions')
ax.legend(fontsize=10, loc='upper left')
ax.axhline(y=0, color='gray', linewidth=0.3)
ax.axvline(x=0, color='gray', linewidth=0.3)
plt.tight_layout()
plt.show()
```

## t-SNE: Non-Linear Projection

t-SNE preserves local neighbourhood structure and often reveals cluster separations that PCA cannot capture in 2D. We test two perplexity values to check robustness.

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 8))

for ax, perp in zip(axes, [10, 30, 60]):
    tsne = TSNE(n_components=2, perplexity=perp, random_state=42, max_iter=1000)
    tsne_result = tsne.fit_transform(X_scaled)

    for i, name in CLUSTER_NAMES.items():
        mask = cluster_labels == i
        ax.scatter(tsne_result[mask, 0], tsne_result[mask, 1],
                   c=colors[i], s=10, alpha=0.5, label=f'{name}')

    ax.set_title(f't-SNE (perplexity={perp})')
    ax.set_xlabel('Component 1')
    ax.set_ylabel('Component 2')
    ax.legend(fontsize=9)

plt.tight_layout()
plt.show()
```

The t-SNE projections confirm that the four K-Means clusters represent genuinely separable groups in
feature space. At perplexity 30, each cluster forms a distinct region with minimal overlap, which the
linear PCA projection (50.7% variance) could not fully capture. The separation remains stable across
perplexity values (10, 30, 60), indicating that the cluster structure is robust and not an artefact of
a specific parameter choice. Plateau Transmission cells form the largest contiguous mass, while Alpine
High-Rise and Urban Vertical produce tightly grouped islands consistent with their specialised profiles.
